Dans quel genre trouve-t-on la plus haute proportion de films provenant du Canada?
Considérez uniquement les films pour lesquels le genre et le pays ne sont pas des valeurs manquantes.

```python
#supprimer les lignes contenant des valeurs manquantes dans les colonnes Genre et Pays
df_sans_na_genre_pays=imdb.dropna(subset=["Genre","Pays"])

# condition  pour filtrer les films provenants du Canada 
condition1=[df_sans_na_genre_pays["Pays"]=="Canada"]

#calculer l'effectif par genre des films provenants du Canada et le reste du monde par un tabeau croisé dynamique, pour compter l'effectif du Canada si
#La condition est vérifée( True), et l'effectif du reste du monde si la condition n'est pas vérifiée( False)

effectif=df_sans_na_genre_pays.pivot_table(values = "Titre",columns=condition1, index = "Genre", aggfunc ="count")


#calculer les proportions par genre des films provenants du Canada(condition est True) par rapport au total par genre
proportions=effectif[True]/effectif.sum(axis=1)

#sélectionner le genre ayant la plus haute proportion de films provenant du Canada
proportion_max=proportions.idxmax()

print("Le genre ayant la plus haute proportion de films provenant du Canada est: {}".format(proportion_max))
```

on considère qu'un film est long à partir d'une durée de 125 minutes. Considérez uniquement les films pour lesquels la durée est connue.

```python

#créer un df, nommé df_durée_connue, en supprimant tous les films avec des durées inconnues.

df_durée_connue=imdb.dropna(subset=["Durée"]) #le parametre subset pour spécifier la colonne ou se trouve les vides

#filtrer les films longs
film_long = df_durée_connue[df_durée_connue["Durée"] >= 120]

#sélectionner les films courts
film_court = df_durée_connue[df_durée_connue["Durée"]<120]

#calculer la difference des votes moyens des filtm longs et les votes moyens des films courts 
différence_votes_moyen=film_long["Votes"].mean()-film_court["Votes"].mean()

# différence de votes moyens arrondie 
np.round(différence_votes_moyen)
```

Le public est-il plus exigeant à l'égard des films ne provenant pas de son marché? Pour répondre à la question, affichez la différence entre la note moyenne des films provenant du même marché que le public et la note moyenne des films ne provenant pas du même marché que le public.

Considérez deux marchés différents : le marché américain et le marché international. Un film est associé au marché américain s'il provient des États-Unis. Un film est associé au marché international s'il ne provient pas des États-Unis. Le public d'un film est associé au marché américain si au moins 25% des votes d'origine connue proviennent des États-Unis. Le public d'un film est international dans le cas contraire. Les films proviennent donc du même marché que le public lorsque les deux catégorisations sont à la fois américaines ou internationales.

Considérez uniquement les films ayant reçu au moins 500 votes, que leur origine soit connue ou non, pour l'analyse.

```python


#1)selectioner les film ayant reçu au moins 500 votes

films_top_votes= imdb[imdb["Votes"] >= 500]



#2)identifier le marché de chaque film dans une nouvelle colonne nommée "Marché", par la fontion marché_film et la methode .apply()
"""
cette fonction retourne le type de marché( NaN,Americain ou Internationl), selon l'élément reçu de la série( ici appliquée à la colonne "Pays").
pour les films d'origine inconnue, le marché est considéré comme inconnu, valeur manquante(NaN).
"""


def marché_film (pays):

    if pd.isna(pays):
        marche = np.nan

    elif pays== "USA":
        marche="Americain"
    else:
        marche="International"
        
    return marche

films_top_votes2=films_top_votes.copy()

films_top_votes2["Marché"]=films_top_votes2["Pays"].apply(marché_film)






#3)identifier le public de chaque film dans une nouvelle colonne nommée "Public", par la fonction public_film et la méthode .apply(lambda row:)

#  Le public d'un film est associé au marché américain si au moins 25% des votes d'origine connue proviennent des USA

# Les votes d'origine connue inclus les votes USA et les votes hors USA(sans les votes des utilisateurs n'ayant pas complété leur origine)

"""
la fonction retourne le type de Public( Americain ou Internationl), selon la proportion 
des votes USA calculée pour chaque film, sur la base des deux l'éléments reçus  des deux 
series(ici appliquée aux colonnes "Votes_hors_USA" et "Votes_USA").
l'utilisation de lmbda row permets de passer les deux arguments de la fonction  pour chaque ligne.
"""

def public_film (vote,votes_USA):

    if (votes_USA/(votes_USA + vote)) >= 0.25:
        public="Americain"
    else:
        public="International"
        
    return public


films_top_votes2["Public"]=films_top_votes2.apply(lambda row:public_film(row["Votes_hors_USA"], row["Votes_USA"]), axis=1)
    
films_top_votes2.head(1)



#4) Calcul des notes moyennes

# 4.1)calculer la notes moyenne des film provenants de même charché que le public
moy1 = films_top_votes2[films_top_votes2["Marché"] == films_top_votes2["Public"]]["Note"].mean()

#4.2)calculer la notes moyennes des film ne provenant pas de même marché que le public

#retirer les notes des films ayant un marché inconnu
films_top_votes3 = films_top_votes2.dropna(subset=["Marché"])

#calculer  la note moyenne ou le marché est different du public
                                        
moy2 = films_top_votes3[films_top_votes3["Marché"] != films_top_votes3["Public"]]["Note"].mean()

#4.3)calculer la différence des deux moyennes
moy1-moy2

```
